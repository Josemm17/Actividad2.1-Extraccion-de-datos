```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv("Rio.csv")
categorias=df.groupby(["host_location"])["host_location"].count()
categorias
```

```python
#Filtro 1 Registros de 5 host name distintos
#Cambie la variable "host_name" por "host_location" 

filtro1=df[(df["host_location"] == "State of Rio de Janeiro, Brazil") | (df["host_location"] == "Rio de Janeiro, Brazil") | (df["host_location"] == "San Diego, CA") | (df["host_location"] == "Brazil") | (df["host_location"] == "Zürich, Switzerland")]
filtro1
```

```python
#Filtro 2 Registros de host que se hayan unido despues del 2020
filtro2=df[df["host_since"]>= "2020-01-01"]
filtro2
```

```python
#Filtro 3 Registros de host que responden a más tardar en un día
filtro3=df[df["host_response_time"]=="within a day"]
filtro3
```

```python
#Filtro 4 Los registros de las filas multiplos de 100
filtro4=df[(df.index)%100==0]
filtro4
```

```python
#Filtro 5 Los registros de los tipos de cuarto entire home y private room

filtro5=df[(df["room_type"] == "Private room") | (df["room_type"] == "Entire home/apt-03-14")]
filtro5
```

```python
#Filtro 6 Los registros que a lo mucho cuenten con 4 camas
filtro6=df[df["beds"]<=4.0]
filtro6
```

```python
#Filtro 7 Los registros que sean superhost
filtro7=df[df["host_is_superhost"]==True]
filtro7
```

```python
#Filtro 8 Los registros de las columnas impares
filtro8=df[(df.index)%2==1]
filtro8
```

```python

filtro1.to_csv("Actividad2.1_filtro1.csv")
filtro2.to_csv("Actividad2.1_filtro2.csv")
filtro3.to_csv("Actividad2.1_filtro3.csv")
filtro4.to_csv("Actividad2.1_filtro4.csv")
filtro5.to_csv("Actividad2.1_filtro5.csv")
filtro6.to_csv("Actividad2.1_filtro6.csv")
filtro7.to_csv("Actividad2.1_filtro7.csv")
filtro8.to_csv("Actividad2.1_filtro8.csv")
```
