```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv("Rio.csv")
df.head(5)
```

```python
#Filtro 1 Los registros de los host cuyo "host_response_rate" sea mayor a 95% y que no sean "superhost"

# Quitar el signo de porcentaje y convertir a nÃºmero
#df['host_response_rate'] = df['host_response_rate'].str.replace('%', '').astype(float)

filtro1=df[~(df["host_is_superhost"] == "t") & (df["host_response_rate"] > 95)]
filtro1
```

```python
#Filtro 2 Los registros cuyas propiedades no tengan 1 baÃ±o
filtro2=df[~(df["bathrooms"]==1)]
filtro2
```

```python
#Filtro 3 Los registros que tengan un precio entre "$100 y $200 US" (Realizar conversiÃ³n de moneda)
tipo_cambio=0.17 #USD A BRL

#Quitar signo de pesos y comas
#df['price'] = df['price'].str.replace('$', '')  
#df['price'] = df['price'].str.replace(',', '')   
# Convertir a nÃºmero
df['price'] = df['price'].astype(float)



df['precio_usd'] = df['price'] * tipo_cambio
df["precio_usd"]

filtro3=df[(df["precio_usd"] >=  100) & (df["precio_usd"] <= 200)]
filtro3
```

```python
#Filtro 4 Los registros que sean "instant_bookable" y que hallan verificado su perfil "host_identity_verified"
filtro4=df[(df["instant_bookable"] ==  "t") & (df["host_identity_verified"] == "t")]
filtro4
```

```python
#Filtro 5 Los registros de los tipos de cuarto: "hotel room" o "Private room"
filtro5=df[(df["room_type"] == "Hotel room") | (df["room_type"] == "Private room")]
filtro5
```

```python
#Filtro 6 Los registros que tengan mas de 4 accommodates
filtro6=df[(df["accommodates"]>4)]
filtro6
```

```python

filtro1.to_csv("Actividad2.2_filtro1.csv")
filtro2.to_csv("Actividad2.2_filtro2.csv")
filtro3.to_csv("Actividad2.2_filtro3.csv")
filtro4.to_csv("Actividad2.2_filtro4.csv")
filtro5.to_csv("Actividad2.2_filtro5.csv")
filtro6.to_csv("Actividad2.2_filtro6.csv")
```
